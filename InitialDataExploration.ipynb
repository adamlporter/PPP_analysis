Drop the last two rows in the table (which don't have GROUP NAME values:

```python
natl = natl.dropna(subset = ['GROUP NAME'])
```

According to this spreadsheet, there are 356,723 congs in the USA. How many are among the HBCs?

```python
hbcs = 0
hbc_list = ['National Baptist','Progressive National','African Methodist Episcopal','Church of God in Christ']
print(f'{"Name":<50}{"N Cong":<10}')
for name in hbc_list:
    #print(name)

    for row in natl[natl['GROUP NAME'].str.contains(name)].iterrows():
        print(f'{row[1][1]:<50} {row[1][2]:<10}') #.row[1][1],row[1][2] )
        hbcs += row[1][2]

print(f'Total number of cong in the BC family {hbcs}.')
```

Why are PNBC double counted? Let's subtract one value and calculate the percent of the total N of cong:

So in the census of congregrations, the 7 BC families comprise about 3.37% of all congregrations in the USA.

Recall that in the PPO data, of the money disbursed to 2,638 (0.29% of the total PPO) religious organizations, 1.85% was disbursed to congs in the BC family.

There maybe innoculous explainations for this disparity (perhaps the BCs have smaller staffs or unpaid ministers) but at least part of the explanation must be the difficulty of applying for PPO funds. To do so required some familarity with the banking industry, legal knowledge helped, etc.
